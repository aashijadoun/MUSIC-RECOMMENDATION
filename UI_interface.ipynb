```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Load your data (replace with your file paths)
spotify_data = pd.read_csv('/content/data.csv')  # Replace 'data.csv' with your dataset file
genre_data = pd.read_csv('/content/data_w_genres.csv')  # Assuming this file contains genre information

# Calculate popularity metrics for songs, artists, and genres
song_popularity = spotify_data.groupby('name')['popularity'].sum().sort_values(ascending=False)
artist_popularity = spotify_data.groupby('artists')['popularity'].sum().sort_values(ascending=False)
genre_popularity = genre_data.groupby('genres')['popularity'].sum().sort_values(ascending=False)

# Function to display top songs by popularity
def display_top_songs(b):
    top_songs_output = widgets.Textarea(value=str(song_popularity.head(10)), disabled=True)
    display(top_songs_output)

# Function to display top artists by popularity
def display_top_artists(b):
    top_artists_output = widgets.Textarea(value=str(artist_popularity.head(10)), disabled=True)
    display(top_artists_output)

# Function to display top genres by popularity
def display_top_genres(b):
    top_genres_output = widgets.Textarea(value=str(genre_popularity.head(10)), disabled=True)
    display(top_genres_output)

# UI for displaying top songs by popularity with expanded size
top_songs_button = widgets.Button(description='Top Songs by Popularity')
top_songs_button.on_click(display_top_songs)
top_songs_button.layout.width = '300px'  # Adjust width
top_songs_button.layout.height = '100px'  # Adjust height

# UI for displaying top artists by popularity with expanded size
top_artists_button = widgets.Button(description='Top Artists by Popularity')
top_artists_button.on_click(display_top_artists)
top_artists_button.layout.width = '300px'  # Adjust width
top_artists_button.layout.height = '100px'  # Adjust height

# UI for displaying top genres by popularity with expanded size
top_genres_button = widgets.Button(description='Top Genres by Popularity')
top_genres_button.on_click(display_top_genres)
top_genres_button.layout.width = '300px'  # Adjust width
top_genres_button.layout.height = '100px'  # Adjust height

# Arrange widgets in a layout
top_songs_ui = widgets.VBox([top_songs_button])
top_artists_ui = widgets.VBox([top_artists_button])
top_genres_ui = widgets.VBox([top_genres_button])

# Display the UI sections
display(top_songs_ui)
display(top_artists_ui)
display(top_genres_ui)
```
